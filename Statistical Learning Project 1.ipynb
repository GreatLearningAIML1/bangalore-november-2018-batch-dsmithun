```python
#Import all relevant libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline

#read csv into df
file_path = "C:/MDS/Trainings/Great lakes/External Lab/16Dec18/Laliga.csv"
teams = pd.read_csv(file_path)
```

```python
teams
```

```python
#replace - with 0
teams = teams.replace(to_replace='-',value=0)
```

```python
#all the teams which have started playing between 1930-1980
teams.dtypes
```

```python
# Convert all number columns from object to int64
teams[["Pos",
      "Seasons",
      "Points",
      "GamesPlayed",
      "GamesWon",
      "GamesDrawn",
      "GamesLost",
      "GoalsFor",
      "GoalsAgainst",
      "Champion",
      "Runner-up",
      "Third",
      "Fourth",
      "Fifth",
      "Sixth",
      "T",
      "BestPosition"]] =            teams[["Pos",
                                            "Seasons",
                                            "Points",
                                            "GamesPlayed",
                                            "GamesWon",
                                            "GamesDrawn",
                                            "GamesLost",
                                            "GoalsFor",
                                            "GoalsAgainst",
                                            "Champion",
                                            "Runner-up",
                                            "Third",
                                            "Fourth",
                                            "Fifth",
                                            "Sixth",
                                            "T",
                                            "BestPosition"]].apply(pd.to_numeric)
```

```python
teams.dtypes
```

```python
# Get debut start by substring
teams["Debut_Start"] = teams["Debut"].str.extract('(....)',expand=True)
teams["Debut_Start"] = pd.to_numeric(teams["Debut_Start"])
```

```python
# Get debut end by substring
teams["Debut_End"] = teams["Debut"].str.slice(5,7)
teams["Debut_End"] = pd.to_numeric(teams["Debut_End"],downcast='integer')
```

```python
teams.dtypes
```

```python
# Store Y for Debut between 1930 and 1980, esle N in Debut 1930-80 column
teams["Debut 1930-80"] = np.where((((teams["Debut_Start"] >= 1930) & (teams["Debut_End"] >= 30))&((teams["Debut_Start"] <= 1980) & (teams["Debut_End"] <= 80))),"Y","N")
```

```python
teams
```

```python
# print all the teams with debut between 1930 and 1980
teams.loc[teams["Debut 1930-80"] == "Y"]
```

```python
#List of teams which came Top 5 in terms of points
teams.sort_values(by="Points",ascending = False,inplace = True)
```

```python
teams.head(5)
```

```python
# Function to return Goal Difference
def Goal_diff_count ():
    teams["Goal Diff"] = teams["GoalsFor"] - teams["GoalsAgainst"]
    return teams["Goal Diff"]
```

```python
Goal_diff_count()
```

```python
#Team with max goal diff
teams.loc[teams["Goal Diff"] == Goal_diff_count().max(),"Team"]
```

```python
#Team with min goal diff
teams.loc[teams["Goal Diff"] == Goal_diff_count().min(),"Team"]
```

```python
teams["Winning Percent"] = (teams["GamesWon"] / teams["GamesPlayed"])*100
teams
```

```python
# Replacing numerical errors with 0
teams.fillna(0,inplace=True)
```

```python
teams
```

```python
teams.sort_values(by="Winning Percent",ascending=False)
```

```python
# top 5 teams which has the highest Winning percentage
teams.head(5)
```

```python
# Aggregate of all positions grouped by Best Position
teams.groupby("BestPosition")["T"].sum()
```

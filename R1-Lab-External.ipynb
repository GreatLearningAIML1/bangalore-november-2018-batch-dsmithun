### Write a function that accepts two numbers as input and prints them - Pass the Sepal length and sepal width of 5th row and print the output

```python
def print_row (sepal_l, sepal_w) :
    output = iris.loc[(iris["Sepal Length (in cm)"] == sepal_l)& (iris["Sepal Width in (cm)"] == sepal_w)]  
    return output
```

```python
print_row (5.0,3.6)
```
